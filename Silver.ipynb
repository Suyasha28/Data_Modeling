APPLY TRANSFORMATIONS & MOVE TO SILVER

```python
%sql
Select *,
upper(customer_name) as Customer_Namr_Upper,
current_date as Process_Date
from datamodeling.bronze.bronze_table
```

```python
spark.sql("""
          Select *,
upper(customer_name) as Customer_Namr_Upper,
current_date as Process_Date
from datamodeling.bronze.bronze_table
""").createOrReplaceTempView("silver_source")
```

###MERGE USING PYTHON

```python
if spark.catalog.tableExists('datamodeling.silver.silvertable'):
    pass
else:
    spark.sql("""create or replace table datamodeling.silver.silvertable
               as 
               select * from silver_source""")
```

```python
%sql
Select * from datamodeling.silver.silvertable
```

### MERGE USING SQL

```python
%sql
MERGE INTO datamodeling.silver.silvertable
using silver_source
on datamodeling.silver.silvertable.order_id = silver_source.order_id
when matched then update set *
when not matched then insert *
```
