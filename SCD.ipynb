### SCD TYPE 1 - UPSERT
### BACK DATA REFRESH CONDITION 
### SOURCE DATE > TARGET DATE

```python
%sql
Merge INTO  datamodeling.gold.scdtype1 as scdtype1
using datamodeling.default.scdtype1_source as source
on scdtype1.Product_id = source.Product_id
when matched and source.processDate >= scdtype1.processDate then update set *
when not matched then insert *
```
